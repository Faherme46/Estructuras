# **Dataset Creation**

```python
# the library is imported to be able to use it later
import pandas as pd
import numpy as np
```

```python
# loading dataset and make a first overview
ruta = "/content/Turismo_Receptivo_en_el_Valle_del_Cauca.csv"
datos = pd.read_csv(ruta)

```

```python
# Create a DataFrame with data
df = pd.DataFrame(datos)

df.info()
```


# **Data selection**

studying the data set, we find that it is good for our purpose. The data is structured, the size is appropriate for the process and is relevant to our mission

In the previous entry we can find that there is not much missing data, this will be quite useful for our purpose.
The colums whith missing data are "Noches" and "Rango edad". Go to review to evaluate if is necesary and possible delete the columns.


We must do a review of the entries

```python
df_ordenado = df.sort_values(by='PAÍS DE PROCEDENCIA')
df_ordenado.head()
```

It can be seen that the department column does not apply to countries other than Colombia, therefore it can be removed from the dataset

```python
try:
  df=df.drop(['DEPARTAMENTO'],axis=1)
except:
  print("Ya fue borrado")
```

```python
#update and review the end of dataset
df_ordenado = df.sort_values(by='PAÍS DE PROCEDENCIA')
df_ordenado.tail(50)
```

We have seen that there are boxes labeled "SIN INFORMACION" which indicates that there is much more missing data than we thought. Therefore, we are going to replace the "SIN INFORMACION" data with NaN entries so that the library can detect them.

```python
#REPLACE DATA AND REVIEW
df=df.replace('SIN INFORMACION', np.nan)
df.info()
```

The PAIS DE PROCEDENCIA  column is not very related to the model we want to develop, therefore we are going to discard this.
We will also discard column MEDIO as its relevance to the model is very low.

```python
#order to delete the columns
try:
  df=df.drop(['MEDIO'],axis=1)    
  df=df.drop(['PAÍS DE PROCEDENCIA'],axis=1)
  
except:
  print("Ya fueron borrados")    

```

We believe that currently gender is not an issue that affects a person's tourist capacity, as well as their reasons for traveling. therefore we will discard the GENERO column

```python
#order to delete the columns
try:
  df=df.drop(['GENERO'],axis=1)    
    
except:
  print("Ya fueron borrados")  
```
